```python
import numpy as np
import matplotlib.pylab as plt
import librosa
import librosa.display 
!git clone https://github.com/jcdevaney/imc2023.git 
```

```python
file = 'imc2023/audioFiles/avm2.wav'
samplingRate = 44100
sig, sr = librosa.load(file,mono=True,sr=samplingRate)
```

```python
plt.plot(sig)
```

```python
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sig)
```

```python
excursion=max(abs(sig))
sigNew = sig / excursion
plt.plot(time,sigNew)
```

```python
plt.plot(time,sigNew)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
```

```python
def readAudioFunction(file, samplingRate):
    sig, sr = librosa.load(file, mono=True, sr=samplingRate)

    time = np.arange(0, sig.shape[0])/sr

    excursion = max(abs(sig))
    sigNew = sig / excursion

    plt.plot(time, sigNew)

    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.show()
file2 = 'imc2023/audioFiles/avm.wav'
readAudioFunction(file2, samplingRate)
```
